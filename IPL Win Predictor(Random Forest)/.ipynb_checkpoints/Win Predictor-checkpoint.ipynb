```python
import numpy as np
import pandas as pd
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
match
```

```python
delivery.head()
```

# we want Batting team,bowling team,city,runs left,balls left,wickets left,total runs,crr,rrr,result

```python
#To Find Total runs scored in each inning we must sum the runs of total runs
total_score_df=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df
```

```python
#WE need only the score of 1st innings so that we can get the score how much they have scored in first batting
total_score_df=total_score_df[total_score_df['inning']==1].reset_index(drop=True)
```

```python
total_score_df
```

```python
# Match-merging is used for merging data sets that have one or more common variables and you want to merge the data sets based on the values of the common variables.
match_df=match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

## this will join the datasets using left_on and right_on if both are same

```python
match_df
```

```python
match_df['team1'].unique()
```

```python
teams=['Sunrisers Hyderabad',
        'Mumbai Indians',
        'Royal Challengers Bangalore',
       'Kolkata Knight Riders',
       'Punjab kings',
       'Chennai Super Kings',
       'Rajasthan Royals',
       'Delhi Capitals'
      ]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Kings XI Punjab','Punjab kings')
match_df['team2']=match_df['team2'].str.replace('Kings XI Punjab','Punjab kings')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

```

## Some of the teams are been removed so we will keep only those teams which are in teams(list)

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

### We must also remove the matches where dls is applied 

```python
match_df['dl_applied']
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df
```

```python
#We Want only

match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
match_df.shape
```

```python
delivery
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
#We want only 2nd innings
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.shape   #These are the total number of balls in 2nd innings
```

```python
delivery_df
```

```python
#We need the score after every bowl
delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()  #cummulative sum
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df
```

```python
# To calculate Number of balls left

delivery_df['balls_left']=126-(delivery_df['over']*6 + delivery_df['ball'])
```

```python
#To calculate wickets left

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")

#lamda (input): (output x) if (condition) true returns x   else (y) returns y

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
# crr = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.describe()  # there are some infinite values this is bcoz there are some runs left after 0 balls
```

```python
final_df=final_df[final_df['balls_left'] !=0]
```

```python
final_df=final_df.sample(frac=1) #shuffling the whole dataframe
```

```python
x=final_df.iloc[:,:-1]
```

```python
y=final_df.iloc[:,9:10]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
x_train.info()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

transformer = make_column_transformer((OneHotEncoder(),['batting_team','bowling_team','city']),remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe= Pipeline(steps=[
    ('step1',transformer),
    ('step2',LogisticRegression())
]
        )
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe= Pipeline(steps=[
    ('step1',transformer),
    ('step2',RandomForestClassifier())
]
        )
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
